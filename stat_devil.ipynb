```python
df = pd.read_csv(f"{path}/heart.csv.xls")
df.head()
```

```python
from sklearn.model_selection import KFold
import numpy as np 
X = df.iloc[:, :-1].to_numpy()
y = df.iloc[:, -1].to_numpy()
```

```python
from sklearn.metrics import confusion_matrix
n_splits = 10  # Number of folds
kf = KFold(n_splits=n_splits)

# Initialize arrays to store coefficients
coef_array = np.zeros((n_splits, X.shape[1]))
cm_array = np.zeros((n_splits, 2, 2))
total_cm = np.zeros((2, 2))

# Perform k-fold cross-validation
fold = 0
for train_index, test_index in kf.split(X):
    # Split data into training and test sets
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Fit logistic regression on training data
    lr = LogisticRegression(max_iter=2000)
    lr.fit(X_train, y_train)
    y_pred = lr.predict(X_test)
    # Store coefficients
    coef_array[fold] = abs(lr.coef_)
    cm = confusion_matrix(y_test, y_pred)
    cm_array[fold] = cm
    total_cm += cm

    # Increment fold
    fold += 1
```

```python
print("Total confusion matrix:\n", total_cm)
```

```python
accuracy = np.sum(np.diag(total_cm)) / np.sum(total_cm)
print("Overall accuracy:", accuracy)
```

```python
sorted_coef_indices = np.argsort(-np.mean(coef_array, axis=0))

plt.boxplot(coef_array[:, sorted_coef_indices])
plt.xticks(range(1, X.shape[1] + 1), [ str(i) for i in df.iloc[:, :-1].columns.to_numpy() ] , rotation=45)
plt.xlabel("Feature")
plt.ylabel("Coefficient")
plt.title("Variance of Coefficients across " + str(n_splits) + "-fold Cross-Validation")
plt.savefig(f"{path}/coef.jpg", dpi=300)
plt.show()
```
